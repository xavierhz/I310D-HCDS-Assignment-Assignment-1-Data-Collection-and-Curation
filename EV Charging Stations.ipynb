The category field seems interesting. What areas are the EV charging stations categorized into?

```python
category_counts = df['Category'].value_counts()
```

```python
category_counts
```

It looks like Education, Healthcare, Retail, Parks and Recreation, and Hospitality are areas that have the least number of 
EV Charging stations within the Plug-In EVerywhere Network. Let's focus on those 

```python
df = df.drop(columns=['Address  1', 'Address  2', 'City', 'State', 'No. of Ports', 'Pricing Policy', 'Usage Access', 'Port 1: Connector Type', 'Voltage (V)', 'Port 2: Connector Type', 'New Georeferenced Column', 'Pricing ', 'Power Select'])
```

```python
def filter_rows_by_values(df, col, values):
    return df[~df[col].isin(values)]
```

```python
df = filter_rows_by_values(df, "Category", ["Workplace","Parking","Multifamily Commercial","Municipal",])
df.sort_values(by=['Category'])
```

These are the EV charging stations remaining after we removed those from categories with a higher count (Workplace, 
Parking, Multifamily Commercial, and Municipal).

Lets recount the number of charging stations per category after removing a few of the higher number categories

```python
category_counts = df['Category'].value_counts()
```

```python
plt.rcParams["figure.figsize"] = (20,10)
category_counts.plot.bar(color=(.8, .4, 0, .7))

plt.title("Number of EV Charging Stations Per Category")
plt.xlabel('Category')
plt.ylabel('Number of EV Charging Stations')
```

This chart shows the number of EV Charging Stations in the different categories which had the least amount. This includes 
Education, Healthcare, Retail, Parks and Recreation, and Hospitality. The City of Austin and Austin Energy may think to
expand their network of electric vehicle charging stations through their GreenChoice program to increase the number of 
charging stations in these categories.
